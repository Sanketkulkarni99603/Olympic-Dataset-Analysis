```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load datasets
athletes = pd.read_csv(r'/content/athlete_events.csv')
regions = pd.read_csv(r'/content/noc_regions.csv')
```

```python
athletes.head()
# prints first 5 rows from dataset 'athlete_events'
```

```python
regions.head()
# prints first 5 rows from dataset 'noc_regions'
```

```python
# Combine dataframes using panda's merge function

athletes_df = athletes.merge(regions, how = 'left', on = 'NOC')
athletes_df.head()
```

```python
athletes_df.shape     #prints no. of rows and columns
```

```python
# To make column names consistent
athletes_df.rename(columns={'region':'Region','notes':'Notes'}, inplace=True);
```

```python
athletes_df.head()
```

```python
athletes_df.info()
```

```python
athletes_df.describe()
```

```python
# To check null values
# True indicates presence of null values
nan_values = athletes_df.isna()
nan_columns = nan_values.any()
nan_columns
```

```python
athletes_df.isnull().sum()
# Gives No. of rows which are null
```

```python
# Teamwise Analysis - India

athletes_df.query('Team == "India"').head(10)
```

```python
# To find most number of participants from countries since start of Olympics
top_20_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(20)
top_20_countries
```

```python
# To plot bar diagram for top 20 countries

plt.figure(figsize=(16,10))
#plt.xticks(rotation=20)
plt.title('Overall participation by country')
sns.barplot(x=top_20_countries.index, y=top_20_countries, palette = 'Set2');
```

```python
# Age distribution of participants

plt.figure(figsize=(16,10))
plt.title('Age distribution of participants at olympics')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age, bins = np.arange(10, 80, 1), color='skyblue', edgecolor = 'white');
```

```python
# Unique sports at Summer Olympics
summer_sports = athletes_df[athletes_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
# Unique sports at Winter Olympics
winter_sports = athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
# Male and female participants

gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# Pie Chart for gender distribution

plt.figure(figsize=(9,5))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',startangle=150, shadow=False);
```

```python
# Total Medals

athletes_df.Medal.value_counts()
```

```python
female_athletes = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex','Year']]
female_athletes = female_athletes.groupby('Year').count().reset_index()
female_athletes.tail(10)
```

```python
womenolympics = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season == 'Summer')]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(25, 12))
sns.countplot(x='Year', data=womenolympics, palette="Spectral")
plt.title('Women Participation')
```

```python
part = womenolympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20, 9))
part.loc[:,'F'].plot()
plt.title('Plot of female participants')
```

```python
# Gold medal athletes
gold_medals = athletes_df[(athletes_df.Medal == 'Gold')]
gold_medals.head()
```

```python
# Avoiding NAN values for gold medals
gold_medals = gold_medals[np.isfinite(gold_medals['Age'])]
```

```python
# Calculate number of athletes having 60+ age
gold_medals['ID'][gold_medals['Age'] > 60].count()
```

```python
# To check type of sports for which above gold medals come
sport_event = gold_medals['Sport'][gold_medals['Age'] > 60]
sport_event
```

```python
# Plot above result
plt.figure(figsize=(12,6))
sns.countplot(sport_event)
plt.title('Gold Medals for Athletes over 60 years')
```

```python
# Country wise gold medal list
gold_medals.Region.value_counts().reset_index(name='Medal').head(10)
```

```python
totalGoldMedals = gold_medals.Region.value_counts().reset_index(name='Medal').head(10)
g = sns.catplot(x="index", y="Medal", data=totalGoldMedals, height=7, kind="bar", palette="rocket")
g.despine(left=True)
g.set_xlabels("Top 10 countries")
g.set_ylabels("Number of Gold medals")
plt.title("Gold medals per country")
```

```python
# Rio Olympics
max_year = athletes_df.Year.max()
print(max_year)

team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == 'Gold')].Team
team_names.value_counts().head(10)
```

```python
# Bar plot for Rio olympics 2016
sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)

plt.ylabel(None);
plt.xlabel('Countriwise Medals in Rio Olympic 2016')
```

```python
not_null_medals = athletes_df[(athletes_df['Height'].notnull()) & (athletes_df['Weight'].notnull())]
```

```python
plt.figure(figsize =(12, 9))
axis = sns.scatterplot(x="Height", y="Weight", data=not_null_medals, hue="Sex")
plt.title('Height vs Weight of Olympic Medalists')
```
